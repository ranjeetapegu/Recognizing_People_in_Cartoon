```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np


df1= pd.read_csv('Cartoon_model_1.csv')
df1['model']='Basic CNN'
df1['Epochs']=df1['Unnamed: 0']
df2= pd.read_csv('Cartoon_model_2.csv')
df2['model']='Data-Augmentation'
df2['Epochs']=df2['Unnamed: 0']
df3= pd.read_csv('Cartoon_model_3.csv')
df3['Epochs']=df3['Unnamed: 0']
df3['model']='Feature Extraction'
df4= pd.read_csv('Cartoon_model_4.csv')
df4['model']='Fine-Tuning'
df4['Epochs']=df4['Unnamed: 0']
frames=[df1,df2,df3,df4]
model_acc= pd.concat(frames)
#model_acc['Epochs']=model_acc['Unnamed: 0']
model_acc[1:10]
```

```python

ax= df1.plot(x='Epochs',y='Acc',color='Orange',label='Basic Model')
df2.plot(x='Epochs',y='Acc',color='DarkBlue',label='Data-Augmentation Model',ax=ax)
df3.plot(x='Epochs',y='Acc',color='DarkGreen',label='Feature Extraction Model',ax=ax)
df4.plot(x='Epochs',y='Acc',color='Red',label='Fine-Tuning Model',ax=ax)
ax.set_title('Training Accuracy')
plt.show()
```

```python

ax= df1.plot(x='Epochs',y='Val_Acc',color='Orange',label='Basic Model')
df2.plot(x='Epochs',y='Val_Acc',color='DarkBlue',label='Data-Augmentation Model',ax=ax)
df3.plot(x='Epochs',y='Val_Acc',color='DarkGreen',label='Feature Extraction Model',ax=ax)
df4.plot(x='Epochs',y='Val_Acc',color='Red',label='Fine-Tuning Model',ax=ax)
ax.set_title('Validation Accuracy')
plt.show()
```

```python
f, ax = plt.subplots(1,1,figsize=(10,6))
df1.plot('Epochs','Val_loss',style='--',color= 'Blue',label='Model1:Val_loss',ax=ax)
df1.plot('Epochs','Loss',marker ="*",color= 'Blue',label='Model1:Loss',ax=ax)
df2.plot('Epochs','Val_loss',style='--',color= 'Orange',label='Model2:Val_loss',ax=ax)
df2.plot('Epochs','Loss',marker ="*",color= 'Orange',label='Model2:Loss',ax=ax)
df3.plot('Epochs','Val_loss',style='--',color= 'Green',label='Model3:Val_loss',ax=ax)
df3.plot('Epochs','Loss',marker ="*",color= 'Green',label='Model3:Loss',ax=ax)
df4.plot('Epochs','Val_loss',style='--',color= 'Red',label='Model4:Val_loss',ax=ax)
df4.plot('Epochs','Loss',marker ="*",color= 'Red',label='Model4:Loss',ax=ax)
ax.set_title('Training vs Validation loss')
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
plt.show()
```
